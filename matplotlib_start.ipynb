## Pyplot API :
#### Univariate : focuses on only one variable, deals with one column, whether it is numerical value or categorical value. 

```python
data = {
    "salary" : [20000, 30000, 25000, 22000, 64000, 34000, 10000, 15000, 32000,] 
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head(5)
```

```python
df.shape
```

```python
# 1.Line Plot :

plt.plot(df["salary"], color= 'red', marker= 'o', linestyle= '--')  # marker will mark the data points on the plot 
plt.grid()
plt.show() 
```

```python
# plotting a histogram :
# Histogram tells us the frequency of the data distributed 

# bins are vertical bars, 
#the height of each bin shows how many data points fall in that range.# plt.grid()

plt.hist(df["salary"], bins= 5, color= 'red')   
plt.show() 
```

### BOX PLOT :

##### 1. A box plot has 2 end points , one representing the minimum value and the other maximum value.
##### 2. Whisker is the line that connects the minimum value to the Lower quartile q1 (25 percentile) i.e 25 percent of the data is below this point.
##### 3. Median is q2 (50t percentile).
##### 4. Upper quartile represents q3(75 percentile). and, there is maximum value .
##### 5. q1 to q3 is interquartile range. 
##### 6. Outliers are the extreme values that are lower than the minimum value OR greater than the maximum value.They are plotted outside the box plot 

```python
# 2. Box Plot :

plt.boxplot(df["salary"])
```

### Univariate : Categorical values 

```python
df["dept"] = ['HR', 'IT', 'HR', 'HR', 'Finance', 'IT', 'Finance', 'IT', 'HR']
```

```python
df.head()
```

```python
df.info()
```

```python
# PIE CHART :

count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels= count.index, autopct= "%1.1f", explode = [0, 0.1, 0]) # explode will seperate the IT sector by 0.1 distance 
plt.show()
```

```python
# COUNT PLOT :

plt.bar(count.index, count , color = ["green", "black", "yellow"])    #count = df["dept"].value_counts()
plt.show()
```

### Bivariate : when we have 2 columns, we analyze the relationship between two columns

##### 1. Numerical - Numerical 

```python
df["Age"] = [22, 25, 28, 30, 35, 39, 45, 50,56]
```

```python
df.head()
```

```python
#1. SCATTER PLOT : analyze direct relationship 

plt.scatter(df["Age"], df["salary"], color = 'orange')
plt.show()
```

```python
sort_age = df.sort_values("Age")
```

```python
# Using Line plot :
plt.plot(sort_age["Age"], df["salary"], color = 'red', marker = 'o')
```

```python
# BAR CHART :

plt.bar(sort_age["Age"], df["salary"], color = 'green')
```
